```python
import pygimli as pg
from pygimli.physics import TravelTimeManager
import matplotlib.pyplot as plt
import numpy as np
import pygimli.meshtools as mt

## 1. load dataset
rstData = pg.DataContainer("../data/tommaso/l1.sgt", "s g")

mgr = TravelTimeManager()

# Define the error in the seismic dataset in sec (e.g. 0.003 means 3 msec)
rstData.set("err", pg.Vector(rstData.size(), 0.002))
# Remove the shoots in the same position of the geophones
rstData.remove(rstData["g"] == rstData["s"])
rstData.remove(rstData["t"] <= 0)
rstData.remove(rstData["t"] > 1)
print(rstData)
rstData.save('../data/tommaso/l1.dat')

# show dataset
fig, ax = pg.plt.subplots(figsize=(8,4))
pg.physics.traveltime.drawFirstPicks(ax, rstData)

ax, cbar = mgr.showData(rstData) # matrix of apparent velocities plt.title(fname)
```

```python
## 2. create mesh
sensors = np.loadtxt("../data/tommaso/geophones_l1.txt")
plc = mt.createParaMeshPLC(sensors, paraDX=.1, boundary=4,
                               #paraDepth=40,
                           paraBoundary=3, paraMaxCellSize=.8)

mesh = mt.createMesh(plc, quality=33, smooth=True)
#mesh.save("mesh_1.bms")
paraDomain = pg.Mesh(2)
paraDomain.createMeshByMarker(mesh, 2)
paraDomain.save("paraDomain_1.bms")
# plot
am, _ = pg.show(paraDomain, alpha=0.3, hold=True, colorBar=False)
pg.viewer.mpl.drawSensors(am, sensors, diam=0.5,
                         facecolor='m', edgecolor='m')
```

```python
## 3. inversion
vMesh = pg.load("paraDomain_1.bms")
mgr.setMesh(vMesh)
inverted = mgr.invert(rstData, mesh=vMesh, #zWeight=0.5,
                      lam=150, vTop=100, vBottom=2500, verbose=True)

print("RST chi:", mgr.inv.chi2())
print("RST rms:", mgr.inv.relrms())

# plot model
ax, cbar = mgr.showResult(logScale=False)
mgr.drawRayPaths(ax=ax, color="k", lw=0.3, alpha=0.5)
ax.set_title(f'$\chi^2$={mgr.inv.chi2():.1f}, '+f'rrmse={mgr.inv.relrms():.1f}', fontsize=10)
```

```python
# 4. plot nice
vmin, vmax = 400, 1800

plt.rcParams['figure.figsize'] = [10, 4.5]
vp , cbar = mgr.showResult(cMap='seismic',cMin=vmin,cMax=vmax,
                           coverage=mgr.standardizedCoverage(),
                           logScale = True, orientation='vertical')
pg.viewer.mpl.drawSensors(vp, sensors, diam=0.5,
                         facecolor='white', edgecolor='black')
vp.set_ylim(-20, 1)
vp.set_xlim(0, 70.5)
#plt.savefig('model_srt.png', dpi=200)
```
